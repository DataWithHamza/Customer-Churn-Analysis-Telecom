## Handling Missing Values and 
## Data Type Conversion

```python
# Check data type of TotalCharges
print(df["TotalCharges"].dtype)

# Count blank values
print((df["TotalCharges"] == " ").sum())
```

```python
# Replace blank spaces with NaN (missing values)
df["TotalCharges"] = df["TotalCharges"].replace(" ", pd.NA)

# Check missing values again
print(df["TotalCharges"].isnull().sum())
```

```python
# Convert TotalCharges to numeric
df["TotalCharges"] = pd.to_numeric (df["TotalCharges"])

# Verify the data type
print(df["TotalCharges"].dtype)
```

```python
print(f"\n{column}")
print(df["TotalCharges"].isnull())
```

```python
# Display tenure for customers with missing TotalCharges
df[df["TotalCharges"].isnull()][["customerID", "tenure", "MonthlyCharges"]]
```

```python
# Remove rows with missing TotalCharges
df = df.dropna(subset=["TotalCharges"])

# Verify
print(df.isnull().sum())
```

```python
# Check duplicate rows
print("Duplicate Rows:", df.duplicated().sum())
```

# SQL Analysis

```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("Telco-Customer-Churn-Cleaned.csv")

df.head()
```

```python
conn = sqlite3.connect("telecom_churn.db")
```

```python
df.to_sql(
    "telecom_churn",
    conn,
    if_exists="replace",
    index=False
)
```

```python
query = """
SELECT *
FROM telecom_churn
LIMIT 5;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS Total_Customers
FROM telecom_churn;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT COUNT(*) AS Churned_Customers
FROM telecom_churn
WHERE Churn = 'Yes';
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
ROUND(
100.0 * SUM(CASE WHEN Churn='Yes' THEN 1 ELSE 0 END)
/ COUNT(*),2
) AS Churn_Rate
FROM telecom_churn;
"""

pd.read_sql(query, conn)
```

```python
def run_query(query):
    return pd.read_sql(query, conn)
```

```python
run_query("""
SELECT
ROUND(AVG(MonthlyCharges),2) AS Average_Monthly_Charges
FROM telecom_churn;
""")
```
